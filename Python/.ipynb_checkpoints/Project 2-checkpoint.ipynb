# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1')
```

```python
df.head(5)
```

```python
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df.dropna(subset = ['InvoiceNo', 'Description'])
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
Remove_C_df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
Remove_C_df
```

# Step 4: Keep Only Positive Quantities

```python
df[df['Quantity'] >= 0]
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower()
df['Description'].str.strip()
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df[df['Country'] == 'United Kingdom']
```

# Step 7: Remove Duplicates

```python
df.drop_duplicates()
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

# Step 9 : Check for Top 10 Selling Products

```python
Top_Selling_Products = df.groupby('Description')
Top_Selling_Products = Top_Selling_Products.InvoiceNo.count()
Top_Selling_Products = Top_Selling_Products.sort_values(ascending=False).head(10)

Top_Selling_Products
```

# Step 10 : Identify Most Active Customers

```python
Most_Active_Customers = df.groupby('CustomerID')
Most_Active_Customers = Most_Active_Customers.InvoiceNo.count() 
Most_Active_Customers = Most_Active_Customers.sort_values(ascending=False).head(10)

Most_Active_Customers
```
