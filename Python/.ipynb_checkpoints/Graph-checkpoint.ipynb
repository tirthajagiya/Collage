```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)  # Plot x and y values
plt.xlabel("X-Axis")  # Label for X-axis
plt.ylabel("Y-Axis")  # Label for Y-axis
plt.title("Simple Line Graph")  # Title of the graph
plt.show()  # Display the graph
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y, color='red', linestyle='--', marker='o', label='Line 1')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Plot Example")
plt.legend()  # Show legend
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.scatter(x, y, color='blue', marker='s', s=100)  # 's' for square markers
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter Plot Example")

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [10, 25, 15, 30]

plt.bar(categories, values, color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

```python
plt.barh(categories, values, color='green')
plt.xlabel("Values")
plt.ylabel("Categories")
plt.title("Horizontal Bar Chart Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generating 1000 random numbers

plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [25, 35, 20, 20]
colors = ["red", "yellow", "pink", "brown"]

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```
