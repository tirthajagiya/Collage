```python
import matplotlib.pyplot as plt
```

## Line Chart

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
y1 = [11,22,33,44,55]
plt.plot(x,y, ls=":", lw=2.5, c="r", marker="*",label = " x vs y")
plt.plot(x, y1, ls="--", lw=2.5 , c="b", marker="o", label = "x vs y1")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("X vs Y")
plt.annotate("Lowest Point", xy=[2.2,19])
plt.legend()
plt.show()
```

### Custom Axis

```python
x = [1,2,3,4,5]
y = [11,22,33,44,55]
ax = plt.axes()
ax.grid()
ax.set_xlim(0,6)
ax.set_ylim(10,60)
ax.set_xticks([0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])
ax.set_yticks([10,15,20,25,30,35,40,45,50,55,60])
plt.plot(x, y, marker="o")
plt.show()
```

## Pie Chart

```python
dept = ["Degree-Eng", "Diploma-Eng", "Management", "Microbiology", "Other"]
stud = [4000, 2500, 1500, 1200, 800]
x = ["r","b","m","g","c"]
plt.pie(stud, labels = dept, colors=x, autopct="%1.2f%%", radius=0.8)
plt.title("Department wise contribution",fontsize=15)
plt.show()
```

### Multiple Graphs on the same ploting canvas

## Bar Chart

```python
mon = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
visiters = [11000, 12000, 15000, 12500, 25000, 17000]
plt.bar(mon, visiters, width=0.5)
plt.show()
```

### Printing the bar height over each bar of the Bar Chart

```python
mon = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
visiters = [11000, 12000, 15000, 12500, 25000, 17000]
bars = plt.bar(mon, visiters, width=0.5)
for i in bars:
    yc = i.get_height()
    plt.text(i.get_x(), yc+100, f"{yc}")
plt.show()
```

### Horizontal Bar Chart

```python
mon = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
visiters = [11000, 12000, 15000, 12500, 25000, 17000]
plt.barh(mon,visiters)
plt.show()
```

### Comparing two values through a bar chart

```python
mon = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
visiters_ggl = [11000, 12000, 15000, 12500, 25000, 17000]
visiters_yahoo = [12000, 11000, 17000, 11800, 20000, 20000]
wd = 0.3
x = range(len(mon))
plt.bar( [i-wd/2 for i in x], visiters_ggl, width=wd, label="Google",color="c")
plt.bar([ i+wd/2 for i in x], visiters_yahoo, width=wd, label ="Yahoo")
plt.xticks(x,mon)
plt.legend()
plt.xlabel("Month")
plt.ylabel("Visiters")
plt.title("Visiters of Google/Yahoo of last 6 Months", color='g', fontsize=15, pad=25)
plt.show()
```

## Scatter Plot

```python
import random
```

```python
random.seed(10)
x = [random.random() for i in range(100)]
y = [random.random() for i in range(100)]
z = [random.random() for i in range(100)]
plt.scatter(x,y, c=z,cmap="plasma")
plt.colorbar()
plt.show()
```

## Histogram

```python
random.seed(5)
age = [random.randint(1,50) for i in range(100)]
plt.hist(age, edgecolor="k", bins=6,histtype="bar")
plt.show()
```

```python
random.seed(5)
age = [random.randint(1,50) for i in range(100)]
plt.hist(age, edgecolor="k", bins=[1,11,21,31,41,51],histtype="stepfilled", color="c")
plt.xticks([1,11,21,31,41,51])
plt.show()
```

## Box Plot

```python
random.seed(10)
time = [ random.randint(20,120) for i in range(100)]
plt.boxplot(time, vert=False)
plt.show()
```

```python
random.seed(10)
time = [ random.randint(20,120) for i in range(100)]
plt.boxplot(time, vert=False, notch=True, widths=0.3)
plt.show()
```

```python
random.seed(10)
time_male = [ random.randint(20,120) for i in range(100)]
time_female = [ random.randint(20,120) for i in range(100)]
plt.boxplot([time_male,time_female], tick_labels=["Male", "Female"])
plt.show()
```
